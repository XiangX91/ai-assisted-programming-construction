This Jupyter Notebook provides a practical, hands-on demonstration of the "Trust but Verify" framework, using the principle of **Incremental Execution** to validate our analysis at every step.


Instead of running a complete script at once, we will execute it in logical chunks. After each step, we will inspect the output to ensure the code is behaving as expected before proceeding. This methodical process builds confidence and helps catch logical errors that an AI might introduceâ€”errors that wouldn't cause the code to crash but would produce incorrect results.

**Step 1: Import Libraries and Load Data**

Our first step is to import the necessary Python libraries and load our dataset. This is our first and most critical verification point: did the data load correctly?

```python
from pathlib import Path
import pandas as pd
import seaborn as sns

# In a Jupyter Notebook, use Path.cwd() to get the current working directory.
# Since this notebook is in the '/notebooks' folder, the project root is its parent.
notebook_dir = Path.cwd()
project_root = notebook_dir.parent

# Construct the full, absolute path to the data file
file_path = project_root / "data" / "Construction-Related_Incidents.csv"

# You can print the path to verify it's correct before using it
print(f"Loading data from the absolute path: {file_path}")

# Now, load the data using this robust, absolute path
try:
    df = pd.read_csv(file_path)
    print("\nDataset loaded successfully.")
except FileNotFoundError:
    print(f"\nError: File not found at the specified path: {file_path}")
    print("Please ensure the '/data' folder and its contents are in the project root directory.")
```

** Verification Point 1: Inspect the Raw Data **

Now that the data is loaded into a pandas DataFrame called df, we must inspect it. We are verifying that:
* The file was read correctly.
* The columns have the expected names (Borough, Incident_Type, etc.).
* The data types are appropriate.

This simple check prevents us from building an entire analysis on incorrectly loaded data.

```python
# Display the first 5 rows of the DataFrame to visually inspect the data
print("First 5 rows of the dataset:")
display(df.head())

# Display a concise summary of the DataFrame, including data types and non-null values
print("\nDataset information:")
df.info()
```

** Step 2: Perform the First Calculation **

Our first analytical step is to count the number of incidents in each borough. Instead of immediately plotting this, we will first perform the calculation and inspect the resulting data structure. This is our second verification point.

```python
# Calculate the number of incidents PER BOROUGH specifically.
# This ensures we get a simple index (a single list of boroughs), not a multi-level one.
borough_counts = df.value_counts()

# --- Second Verification Point ---
# Print the result of our calculation to verify it before plotting.
# The output should now be a simple list of boroughs and their counts.
print("Calculated incident counts per borough:")
print(borough_counts)
```

** Step 3: Generate the First Visualisation ** 

Having verified that our borough_counts calculation is correct, we can now proceed with confidence to the visualisation step. We trust the input to this step because we have just verified it.

```python
import matplotlib.pyplot as plt

# Set a professional and clean style for the plots.
sns.set_style("whitegrid")
    
# Calculate the number of incidents per borough and sort them in descending order.
borough_counts = df['Borough'].value_counts().sort_values(ascending=False)
    
# Create the figure and axes for the plot.
plt.figure(figsize=(12, 7))
# CORRECTED: Removed the 'hue' parameter as it was redundant and caused the error.
ax1 = sns.barplot(x=borough_counts.index, y=borough_counts.values, palette='viridis')
    
# Set a clear title and labels for the axes.
ax1.set_title('Total Construction Incidents by NYC Borough', fontsize=16, fontweight='bold')
ax1.set_xlabel('Borough', fontsize=12)
ax1.set_ylabel('Total Number of Incidents', fontsize=12)
    
# Add numerical counts on top of each bar for better readability.
for p in ax1.patches:
    ax1.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 9), 
                    textcoords='offset points',
                    fontsize=11)
    
# Rotate the x-axis labels to prevent them from overlapping.
plt.xticks(rotation=45, ha='right')
# Ensure all plot elements fit nicely within the figure.
plt.tight_layout()
# Display the first plot.
plt.show()
```

** Step 4: Prepare Data for the Second, More Complex Visualisation **

Our next goal is to create a stacked bar chart showing the breakdown of incident types within each borough. This requires more complex data manipulation: identifying the top incident types and creating a pivot table. This is a critical stage where logical errors can easily occur.

Verification Point 3: Inspect the Pivot Table
Before plotting, we will display the pivot table. We are verifying that:
* The rows are the boroughs.
* The columns are the top 5 incident types.
* The values are the counts of incidents.

```python
if 'Check2 Description' in df.columns:
    df.rename(columns={'Check2 Description': 'Incident_Type'}, inplace=True)

# Identify the top 5 most common incident types to keep the visualisation clean
top_incident_types = df['Incident_Type'].value_counts().nlargest(5).index

# Filter the DataFrame to only include these top incident types
df_top_incidents = df[df['Incident_Type'].isin(top_incident_types)]

# Create a pivot table to count incident types per borough.
# We must tell groupby which columns to use: 'Borough' and 'Incident_Type'.
incident_pivot = df_top_incidents.groupby(['Borough', 'Incident_Type']).size().unstack(fill_value=0)

# Reorder the pivot table based on the total incidents per borough for better visualisation
incident_pivot = incident_pivot.loc[borough_counts.index]

# --- Third Verification Point ---
print("Pivot table of top 5 incident types per borough:")
display(incident_pivot)
```

** Step 5: Generate the Final Visualization **

Having confirmed that our pivot table is structured correctly, we can now confidently create our final, more complex visualization. This incremental process ensures that any errors would have been caught in the preceding steps, making our final output much more reliable.

```python
# Create the stacked bar chart from our verified pivot table
incident_pivot.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='plasma')

# Add titles and labels
plt.title('Breakdown of Top 5 Incident Types by NYC Borough', fontsize=16, fontweight='bold')
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Number of Incidents', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Incident Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.tight_layout()
plt.show()
```

** Conclusion **

By executing the code incrementally and verifying the output at each logical step, we have built a high degree of trust in our final analysis. This process directly addresses the risks of AI-generated code by applying human oversight and domain knowledge at critical checkpoints, perfectly illustrating the "Trust but Verify" mindset.
